```python
import pandas as pd
import sqlite3
```

```python
# Read sqlite query results into a pandas DataFrame
con = sqlite3.connect("imdb.sqlite")
ratings_df = pd.read_sql_query("SELECT * from ratings", con)
movies_df = pd.read_sql_query("SELECT * from movies", con)

names_df = pd.read_sql_query("SELECT * from names", con)

# Verify that result of SQL query is stored in the dataframe
print(movies_df.head())

con.close()
len(names_df)
```

```python
inner_join_df= pd.merge(movies_df, ratings_df, on='tconst', how='inner')
```

```python

inner_join_df
inner_join_df.to_csv("IMDBmovieRatings.csv", sep='@', encoding='utf-8')
names_df.to_csv("IMDBnames.csv", sep='@', encoding='utf-8')
```
